```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("D:\\Cursos\\Analitica de datos\\base_datos_2008.csv",nrows =100000)
data = np.unique(df.DayOfWeek, return_counts=True)
labs= ["Lun","Mar","Mie","Jue","Vie","Sab","Dom"]
data
```

```python
plt.pie(x=data[1], 
    labels=labs,
    radius=.7,
    colors=sns.color_palette("hls",7), #["Red","Green","Orange","Blue","Gray","Pink","Black"]labeldistance = 0
    explode = (0,0,0.2,0,0,0,.1),#saca las tortas
    startangle=90,
    autopct = "%1.1f%%" )

plt.legend(loc = "best",labels = labs)   
plt.show()
```

```python
plt = sns.barplot(x = labs, y = data[1])
plt.set(xlabel ="Dia de la semana", ylabel ="Numero de vuelos")
```
