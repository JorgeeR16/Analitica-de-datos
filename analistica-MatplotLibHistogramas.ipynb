```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("D:\\Cursos\\Analitica de datos\\base_datos_2008.csv",nrows =1000000)
```

```python
df.dropna(inplace=True,subset= ["ArrDelay","DepDelay","Distance"])
```

```python
sns.distplot(df["Distance"], kde = False, bins=100)
```

```python
sns.kdeplot(df["ArrDelay"])
sns.kdeplot(df["DepDelay"])
plt.xlim([-300,300])
```
